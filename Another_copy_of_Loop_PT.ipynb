**Practice in Loop**


1. Write a for loop that prints numbers from 1 to 10.

```python
for i in range(1, 11):
  print(i)
```

2. Write a while loop that prints “Hello” five times.

```python
i = 0
while i < 5:
  print(i)
  i += 1
```

3. Write a loop that prints only even numbers from 1 to 20 using continue.

```python
for i in range(2, 21):
  if i % 2 != 0:
    continue
  print(i)
```

4. Create a nested loop that prints a 3×3 square of stars (*).
Example output:

' * * *

' * * *

' * * *

```python
for i in range(3):
  print("*" * 3)
```

5. Use break in a loop to stop when the number is 7 (loop from 1 to 10).

```python
for i in range(1, 11):
  if i == 7:
    break
  print(i)
```

6. Sum of Numbers (For Loop)

Write a program that calculates the sum of numbers from 1 to 50.

```python
result = 0
for i in range(1, 51):
  result += i
print(f"The sum of numbers from 1 to 50 is: {result}")

```

```python
for i in range(5):
  for j in range(i+1):
    print("*", end =" ")
  print()
```

6. Password Checker (While Loop)

Let the user enter a password.
Keep asking until the user types "admin123".

**Output:**
Access granted!


```python
password = " "
while password != "admin123":
  password = input("Enter a password: ")
print("Access granted!")
```

7. Multiplication Table (Nested Loop)

Write a program to print the 1–5 multiplication tables.

Example:
1 x 1 = 1
1 x 2 = 2
…
5 x 10 = 50

```python
for i in range(1, 6):
  print(f"\nmultiplication Table of {i}:\n")
  for j in range(1,11):
    print(f"{i} x {j} = {i * j}")
```


8. Skip Multiples of 3 (Continue)

Print numbers from 1 to 30, but skip numbers divisible by 3.

```python
for i in range(1, 31):
  if i % 3 == 0:
    continue
  print(i)

```
